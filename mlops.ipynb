# Edit distance 

```python
s1 = "prayer"
s2 = "repay"

def EditDistance(s1,s2):
    D = [[0 for j in range(len(s2)+1)] for i in range(len(s1)+1)]
    for i in range(1,len(s1)+1):
        D[i][0]= D[i-1][0]+1
    for j in range(1,len(s2)+1):
        D[0][j]= D[0][j-1]+1
    for i in range(1,len(s1)+1):
        for j in range(1,len(s2)+1):
            d=D[i-1][j]+1
            a=D[i][j-1]+1
            s=D[i-1][j-1] + (s1[i-1] != s2[j-1]) * 1
            D[i][j]=min(d,a,s)
    return D[len(s1)][len(s2)]       
EditDistance(s1,s2)            
```

# Optimal Edit distance 

```python
class cell:
    def __init__(self, val):
        self.value = val
        self.parent = []

    def __add__(self, other):
        return cell(self.value + other)

    def __lt__(self, other):
        return self.value < other.value


def optimal_editdistance(s1, s2):
    D = [[cell(0) for i in range(len(s2)+1)] for j in range(len(s1)+1)] 

    for i in range(1, len(s1)+1):
        D[i][0] = D[i-1][0] + 1
        D[i][0].parent.append((i-1, 0))

    for j in range(1, len(s2)+1):
        D[0][j] = D[0][j-1] + 1
        D[0][j].parent.append((0, j-1))

    for i in range(1, len(s1)+1):
        for j in range(1, len(s2)+1):
            d = D[i-1][j] + 1
            a = D[i][j-1] + 1
            s = D[i-1][j-1] + (s1[i-1] != s2[j-1]) * 1
            D[i][j] = min(d, a, s)
            

            if D[i][j].value == d.value:
                D[i][j].parent.append((i-1, j))
            if D[i][j].value == a.value:
                D[i][j].parent.append((i, j-1))
            if D[i][j].value == s.value:
                D[i][j].parent.append((i-1, j-1))

    return D[len(s1)][len(s2)].value,D



distance,D = optimal_editdistance(s1,s2)
print("The edit distance is: ", distance)
c = D[len(s1)][len(s2)]
while len(c.parent) > 0:
    print(c.parent)
    c = D[c.parent[0][0]][c.parent[0][1]]
    print(c.value)
```

# Byte pair encoding 

```python
from collections import defaultdict
import re

def get_frequent_pairs(vocab):
    pairs = defaultdict(int)
    for word, freq in vocab.items():
        alphabets = word.split('_')
        for i in range(len(alphabets) - 1):
            pairs[(alphabets[i], alphabets[i + 1])] += freq  
    return pairs        

def merge_frequent_pairs(pair, vocab):
    joining = re.escape('_'.join(pair))
    new_vocab = {}
    for word, freq in vocab.items():
        new_word = re.sub(joining, ''.join(pair), word)
        new_vocab[new_word] = freq
    return new_vocab

vocab = {'s_l_o_w_<w>': 2, 's_l_o_w_e_r<w>': 5, 'l_o_w_e_s_t_<w>': 1, 'p_r_o_v_i_d_e_r_<w>': 3, 'r_i_d_e_r_<w>': 4}
num_merges = 11

for _ in range(num_merges):
    pairs = get_frequent_pairs(vocab)
    best_pair = max(pairs, key=pairs.get)
    vocab = merge_frequent_pairs(best_pair, vocab)
    print(f"Merge: {best_pair} -> {vocab}")
```
